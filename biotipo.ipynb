```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns

df = pd.read_csv('dados_biotipo.csv',na_values=' ')
df
```

```python
#  Peso

x = df['IMC'] * (df['Altura'] * df['Altura'])
df['Peso'].fillna(round(x,2), inplace=True)
df
```

```python
#  Altura

y = np.sqrt(df['Peso'] / df['IMC'])
df['Altura'].fillna(round(y,2), inplace=True)
df
```

```python
# Cor dos Olhos

df['Cor Olho'].fillna(df['Cor Olho'].mode()[0], inplace = True)
df
```

```python
# Cor do Cabelo


df['Cor Cabelo'].fillna(df['Cor Cabelo'].mode()[0], inplace = True)
df
```

```python
# Outlier

df['Outlier'] = df['IMC'].apply(lambda z: '1' if z < 18.5 or z > 40 else '0')
sns.scatterplot(x=df['Peso'], y=df['Altura'], hue=df['Outlier'], palette={'0': 'blue', '1': 'red'})
df = df.sort_values(by='IMC', ascending=False)
plot.show()
```

```python
df = df.drop(df[df['Outlier'] == '1'].index)
df.reset_index(drop=True, inplace=True)
df = df.drop('Outlier', axis=1)  # 0 = linha, 1 = coluna
df
```

```python
df.to_csv('dados_biotipo.csv', index=False)
```

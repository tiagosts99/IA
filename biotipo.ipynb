```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
from scipy import stats
import sweetviz as sw

df = pd.read_csv('dados_biotipo.csv',na_values=' ')
df
```

```python
#  Peso

x = df['IMC'] * (df['Altura'] * df['Altura'])
df['Peso'].fillna(round(x,2), inplace=True)
df
```

```python
#  Altura

y = np.sqrt(df['Peso'] / df['IMC'])
df['Altura'].fillna(round(y,2), inplace=True)
df
```

```python
# Cor dos Olhos

df['Cor Olho'].fillna(df['Cor Olho'].mode()[0], inplace = True)
df
```

```python
# Cor do Cabelo


df['Cor Cabelo'].fillna(df['Cor Cabelo'].mode()[0], inplace = True)
df
```

```python
# Outlier

df['z_scores'] = stats.zscore(df['IMC'])
df['A'] = df['z_scores'].apply(lambda z: '1' if z > 3 else '0')
df = df.sort_values(by='A', ascending=False)
df = df.drop(df[df['A'] == '1'].index)

```

```python
def __init__ ()

df= sw.analyze(df)
df.show.html()
```
